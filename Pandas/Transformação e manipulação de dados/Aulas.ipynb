# Aula 1: Entendendo o problema

```python
# Conhecendo o problema
import pandas as pd

dados = pd.read_json('dados_hospedagem.json')
dados.head()
```

```python
dados = pd.json_normalize(dados['info_moveis']) # pd.json_normalize(dados['info_moveis']) caso os dados estejam aninhados como é o caso utilizamos o normalize para nos retornar no formado de tabela com multiplas colunas
dados.head() # vamos repara que as colunas armazenam dados agrupados em listas
```

# Aula 2: Dados Numericos

Oque cada coluna da tabela armazena

avaliacao_geral: refere-se à média de notas dadas para a avaliação da hospedagem no imóvel.

experiencia_local: descreve as experiências oferecidas durante a hospedagem no imóvel.

max_hospedes: informa a quantidade máxima de hóspedes que o local permite.

descricao_local: descreve o imóvel.

descricao_vizinhanca: descreve a vizinhança ao redor do imóvel.

quantidade_banheiros: informa a quantidade de banheiros disponíveis.

quantidade_quartos: informa a quantidade de quartos disponíveis.

quantidade_camas: informa a quantidade de camas disponíveis.

modelo_cama: informa o modelo de cama oferecido.

comodidades: informa as comodidades oferecidas pelo imóvel.

taxa_deposito: informa a taxa de depósito mínima para segurança de hospedagem.

taxa_limpeza: informa a taxa cobrada para o serviço de limpeza.

preco: refere-se ao preço base a ser cobrado pela diária no imóvel.

```python
# Retirando as listas

colunas = list(dados.columns) # pega o nome das colunas e salva elas em uma lista e salva a lista dentro da variavel colunas
colunas
```

```python
# como vimos na tabela só precisamos desagrupas os dados que estão em listas apartir da coluna descricao_local, pois os anteriores já estõa normais 

dados = dados.explode(colunas[3:]) # o metodo expode separa os dados que estão em listas das colunas que ele recebe como parametro, o index não muda precisando ser resatado após o metodo explode
dados.head()
```

```python
dados.reset_index(inplace=True, drop = True) # inplace = True diz para fazer a mudança dentro do DF(se não for usado ele não muda direto no DF), drop= True não cria uma coluna com os indexes antigos do DF
dados.head()
```

```python
dados.info()
```

```python
# Convertendo dados Numéricos

import numpy as np

# astype(tipo de dados) converte tipo de dados se for possivel de uma coluna
dados['max_hospedes'] = dados['max_hospedes'].astype(np.int64) # int64 armazena numeros inteiros com no maximo 64 bits
dados.info()
```

```python
col_numericas = ['quantidade_banheiros','quantidade_quartos','quantidade_camas'] # colunas que quer alterar o tipo para int

dados[col_numericas] = dados[col_numericas].astype(np.int64)

dados.info()
```

```python
dados['avaliacao_geral'] = dados['avaliacao_geral'].astype(np.float64)
dados.info()
```

caso tente converter uma coluna par aum tipo de dados que ela não possui ela da erro, EX: tentar converter a coluna preco pra float, ela vai dar erro pois temos o "R$" que é do tipo str

```python
# Numeros em strings

dados['preco'] = dados['preco'].apply(lambda x: x.replace('$', '').replace(',', '').strip()) #aplly aplica algo em algo no caso aplica a função que está dentro dos parenteses na coluna preço linha por linha, replace substitui caracteres de uma string no caso '$' por '' ('' = nada), strip apga espaços em branco antes e depois das strings, resumo: aqui formatamos o dado do tipo object tirando o '$' a ',' e os espaços vazios
```

```python
dados['preco'] = dados['preco'].astype(np.float64)

dados.info()
```

```python
dados[['taxa_deposito','taxa_limpeza']] = dados[['taxa_deposito','taxa_limpeza']].applymap(lambda x: x.replace('$', '').replace(',', '').strip()) #applymap aplica a função em dataframes
```

```python
dados[['taxa_deposito','taxa_limpeza']] = dados[['taxa_deposito','taxa_limpeza']].astype(np.float64)

dados.info()
```

# Aula 3: Dados textuais

Vamos fazer uma tokenização simples

```python
# O problema do texto

dados['descricao_local'] = dados['descricao_local'].str.lower() # trsformas os dados da coluna em string e todas em letras minusculas
dados.head()
```

```python
# Removendo caracteres com regex

dados['descricao_local'][3169]
```

```python
dados['descricao_local'] = dados['descricao_local'].str.replace('[^a-zA-Z0-9\-\']',' ', regex=True) # [^a-zA-Z0-9\-\'] ^ siguinifica negação, no caso acima ele está negando a seleção de caracteres 'como se disse não mude esses caracteres que te passei, os outros troque por ' ' (espaço vazio)
```

```python
dados['descricao_local'] = dados['descricao_local'].str.replace('(?<!\w)-(?!\w)',' ', regex=True) # (?<!\w)-(?!\w) verifica se não há caracter antes e depois do ifen, se não existir caracter antes e depois do ifen ele vai ser selecionado pelo replace 
```

```python
# Tokenização de strings

dados['descricao_local'] = dados['descricao_local'].str.split() # split() divide os elementos textuais em uma lista usando o espaço em branco como o separador
dados.head()
```

```python
dados['comodidades'] = dados['comodidades'].str.replace('\{|}|\"', '', regex=True) # esse regex está procrando { ou } ou " e substitui por um espaço vazio
```

```python
dados['comodidades'] = dados['comodidades'].str.split(',') # dentro do split vai o separador oque separa cada coisa, caso não seja passado ele vai usar como separador espaços 
dados.head()
```

```python
# limpando a coluna descrição vizinhança

dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.lower() # pega todas as letras e trasnformas elas em minusculas
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex=True) # o regex seleciona todos os caracteres que não sejam letras maiusculas e minusculas, numeros e ', todos que não sejam esses caracteres vão ser substituidos por um espaço
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.replace('(?<!\w)-(?!\w)', '', regex=True) # esse regex verifica se tem caracteres antes e depois do ifen se tiver não remove o ifen se não tiver substitui o ifen por um espaço vazio 
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.split() # separa os tokens por espaços e salva eles em uma lista
dados.head()
```
